```python
import pandas as pd

"""
Adding a query column to the dataset in order to search for the songs
in the Spotify API and grab their popularity
"""
df = pd.read_csv('data/processed/spotify_lyrics_features.csv')
df['query'] = df['song'] + ' ' + df['artist']
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time

#Initiating a connection with the Spotify API
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id='ec23a230eb11474dab3b1542018a4db2',
    client_secret='472ed4aa3f2e487a8045f4eac21d711f'
))

#Helper function to extract the popularity of each song
def get_pop(query):
    res = sp.search(query, type='track', limit=1)
    items = res['tracks']['items']
    if len(items) > 0:
        return items[0]['popularity']
    return None

pops = []

for i, q in enumerate(df["query"]):
    print(f"Fetching: {q} ({i+1}/{len(df)})")
    try:
        pop = get_pop(q)
    except Exception as e:
        print("Error:", e)
        pop = None
    
    pops.append(pop)
    time.sleep(0.12)

df["popularity"] = pops
```

```python
df.to_csv('data/processed/spotify_lyrics_features_sampled_pop.csv', index=False)
```
